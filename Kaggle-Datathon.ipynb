```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

## "Cinsiyet" sütununu kategorik bir değişkenden sayısal bir değişkene dönüştürmek

```python
 train_data["Cinsiyet"] = train_data['Cinsiyet'].replace(["Erkek"], 1)
train_data["Cinsiyet"] = train_data['Cinsiyet'].replace(["Kadın"], 0)
```

```python
train_data
```

## Bir veri çerçevesindeki kategorik değerleri numerik değerlere dönüştürmek

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()


train_data['Öbek İsmi'] = label_encoder.fit_transform(train_data['Öbek İsmi'])
```

```python
train_data = pd.get_dummies(train_data, columns=['Yaş Grubu','Medeni Durum','Eğitim Düzeyi','Yaşadığı Şehir', 'İstihdam Durumu','Eğitime Devam Etme Durumu', 'En Çok İlgilendiği Ürün Grubu'])
train_data
```

```python
test_data["Cinsiyet"] =test_data["Cinsiyet"].replace(["Erkek"],1)
test_data["Cinsiyet"] =test_data["Cinsiyet"].replace(["Kadın"],0)
```

## Belirlenen özellikleri modelin anlayabileceği bir sayısal forma getirmek

```python
 
test_data = pd.get_dummies(test_data, columns=['Yaş Grubu','Medeni Durum','Eğitim Düzeyi','Yaşadığı Şehir', 'İstihdam Durumu','Eğitime Devam Etme Durumu', 'En Çok İlgilendiği Ürün Grubu'])
test_data
```

## Makine öğrenimi modeli eğitmi ve test etme işlemleri

```python
X = train_data.drop("Öbek İsmi", axis=1)  # "target" sütunu bağımlı değişkeni temsil ediyor
y = train_data["Öbek İsmi"]

# Eğitim ve test verilerini bölelim
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.36, random_state=39)


model = RandomForestClassifier(n_estimators=50, random_state=39)
model = RandomForestClassifier(n_estimators=50, max_depth=12, random_state=39)
model = RandomForestClassifier(n_estimators=50, min_samples_leaf=8, random_state=39)





model.fit(X_train, y_train)


# Modeli test verileriyle değerlendirilir
y_pred = model.predict(X_test)

# Doğruluk değerini hesaplanır
accuracy = accuracy_score(y_test, y_pred)
print("Model Doğruluğu:", accuracy)

# Şimdi, test verileri üzerinde tahminler yapabilirsiniz:
test_tahminleri = model.predict(test_data)
```

```python
test_tahminleri = test_tahminleri+1
```

## Oluşturulan tahminler için veri çerçevesi oluşturma

```python
veri = {
    'id': range(2340),
    'Öbek İsmi': test_tahminleri
}
Tahmin_Dataframe = pd.DataFrame(veri)
Tahmin_Dataframe['Öbek İsmi'] = Tahmin_Dataframe['Öbek İsmi'].apply(lambda test_tahminleri: 'obek_' + str(test_tahminleri))
Tahmin_Dataframe.set_index('id', inplace=True)
Tahmin_Dataframe
```

```python
Tahmin_Dataframe.to_csv('submission.csv')
```
